```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equations
def model(y, t, beta, k, d, dei2, p, c, pc, cc, kc, ccl, ke, de, pel, rel, Kel):
    U, I1, I2, A, V, C, CL, E, EL = y
    
    # Define the equations based on the LaTeX model
    dU_dt = -beta * U * V
    dI1_dt = beta * U * V - k * I1
    dI2_dt = k * I1 - (d + dei2 * E) * I2
    dA_dt = (d + dei2 * E) * I2
    dV_dt = p * I2 - c * V
    dC_dt = pc * (I1 + I2) - cc * C - kc * C
    dCL_dt = kc * C - ccl * CL
    dE_dt = ke * EL - de * E
    dEL_dt = pel * CL + rel * CL * EL * Kel / (Kel + EL) - ke * EL
    
    return [dU_dt, dI1_dt, dI2_dt, dA_dt, dV_dt, dC_dt, dCL_dt, dE_dt, dEL_dt]

# Initial conditions
U0 = 1440  # Initial value for U
I1_0 = 160  # Initial value for I1
I2_0 = .0  # Initial value for I2
A0 = 0.0  # Initial value for A
V0 = .0  # Initial value for V
C0 = 0.0  # Initial value for C
CL0 = 0.0  # Initial value for CL
E0 = 0.0  # Initial value for E
EL0 = 0.0  # Initial value for EL

# Combine initial conditions into a list
initial_conditions = [U0, I1_0, I2_0, A0, V0, C0, CL0, E0, EL0]

# Time points (0 to 50 with 1000 intervals)
t = np.linspace(0, 7200, 14400)

# Parameters (you need to adjust these as needed)
beta = 7.18*60*1e-8
k = 1.85*60*1e-5
d = 2.78*60*1e-6
dei2 = 6.96*1e-11
p = 1.09*60*1e-4
c = 1.09*60*1e-4
pc = 6.96 * 1e-3
cc = 1.386 * 1e-3
kc = 5.79*60*1e-6
ccl = 5.79*60*1e-6
ke = 1.16*1e-5*60
de = 1.16*1e-5*60
pel = 1.16*1e-9*60
rel = 5.79*60*1e-8
Kel = 500

# Solve ODEs
solution = odeint(model, initial_conditions, t, args=(beta, k, d, dei2, p, c, pc, cc, kc, ccl, ke, de, pel, rel, Kel))

# Extract each variable for plotting
U = solution[:, 0]
I1 = solution[:, 1]
I2 = solution[:, 2]
A = solution[:, 3]
V = solution[:, 4]
C = solution[:, 5]
CL = solution[:, 6]
E = solution[:, 7]
EL = solution[:, 8]

# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(t, U, label='U (Uninfected)')
plt.plot(t, I1, label='I1 (Infected)')
plt.plot(t, I2, label='I2 (Infected producing Virus)')
plt.plot(t, A, label='A (Dead cells)')
plt.plot(t, V, label='V (Virus)')
plt.plot(t, C, label='C (Cytokine Level Local)')
plt.plot(t, CL, label='CL (Cytokine Level Lymph Node)')
plt.plot(t, E, label='E (Immune Cells Local)')
plt.plot(t, EL, label='EL (Immune Cells Lymph Node)')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.grid(True)
plt.show()

print(EL)
```
